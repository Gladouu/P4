### Répartition du CA par catégorie

```python
# First data pie needed

cat0 = join.loc[join['categ'] == 0]
cat1 = join.loc[join['categ'] == 1]
cat2 = join.loc[join['categ'] == 2]

pie0 = cat0['price'].sum()
pie1 = cat1['price'].sum()
pie2 = cat2['price'].sum()


# Second data pie needed

sell = join.groupby('categ')['session_id'].count()


# First Pie

plt.figure(figsize=(20, 12))

plt.subplot(2, 2, 1)

labels=['Catégorie 0', 'Catégorie 1', 'Catégorie 2']

plt.pie((pie0, pie1, pie2), labels=labels, autopct='%1.f%%')
plt.title('Répartition du CA par catégorie (12 mois)', fontsize=18, pad=30)


# Second Pie

plt.subplot(2, 2, 2)

labels=['Catégorie 0', 'Catégorie 1', 'Catégorie 2']

plt.pie(sell, labels=labels, autopct='%1.f%%')
plt.title('Répartition des ventes par catégorie (12 mois)', fontsize=18, pad=30)

#plt.savefig('04. CA & ventes par catégorie - Octobre 2021 - pie.jpg', dpi=300)
plt.show()
```

### Top 10 et Outliers

```python
# Top 10 clients 

join.groupby('client_id')['price'].sum().sort_values(ascending=False).head(10)
```

```python
# On constate ici qu'il y a 4 clients qui se démarquent par le montant de leurs achats sur l'année. 

# A priori, il semblerait que cela soit des valeurs atypiques : elles sortent du lot sans pour autant être 
# fausses. Plusieurs hypothèses peuvent être émisent au sujet de ses clients : bibliothèques, revendeurs, écoles,
# (B2B).

# Il faudra donc être prudent en fonction des calculs réalisés.
```

```python
# Ajout des colonnes âge et tranche d'âge pour la suite de l'analyse 

join['age'] = 2022 - join['birth']

# Les tranches d'âge ont été déterminées à postériori en fonction du graphique dans la section 4.3 (graphique 17) 
# mais misent ici pour une question pratique. 

def tranche_age(x) : 
    if x<=30 : return('18-30')
    elif x<=50 : return('31-50')
    elif x<=70 : return('51-70')
    else : return('71+') 
    
join['tranche_age']=join['age'].apply(tranche_age)


# Exclusion des clients considérés comme des outliers par rapport à la somme achetée sur l'année :

# Join Without Outliers Clients

joinWO =   join[~(join['client_id'] == 'c_1609')
                ^ (join['client_id'] == 'c_4958' )
                ^ (join['client_id'] == 'c_6714' )
                ^ (join['client_id'] == 'c_3454' )]


# DF with only Outliers Clients

c_outliers = join[(join['client_id'] == 'c_1609')
                ^ (join['client_id'] == 'c_4958' )
                ^ (join['client_id'] == 'c_6714' )
                ^ (join['client_id'] == 'c_3454' )]
```

#### Hypothèse H0

```python
# Pour savoir s'il existe une corrélation entre deux variables qualitatives, on réalise le test d'indépendance
# du Chi2. 

# Pour cela, on détermine deux hypothèses :
```

$\left\{
\begin{array}{l}
 H_0\ : \ les\ deux\ variables\ étudiées\ sont\ indépendantes \\
 H_1\ : \ les\ deux\ variables\ étudiées\ sont\ dépendantes
\end{array}

#### Création du tableau de contingence

```python
cont = joinWO[['sex','categ']].pivot_table(index='sex', columns='categ', aggfunc=len)
# penser à .fillna(0) le cas échéant

chi2, p, dof, exp = st.chi2_contingency(cont)

cont
```

#### Test du Chi2

```python
# Risque acceptable, de rejeter H0 alors qu'elle est vraie
print("Risque d'erreur α = 0.05") 

print("Degré de liberté =", dof)
print("")

print("Chi2 =", round(chi2,3))

# Table des valeurs critiques de la loi du Chi2 
# Valeur qui ne doit pas être dépassée pour ne pas rejeter l'hypothèse nulle
# Pour un dof = 2 et un alpha 5%, Chi2 table = 5,991 
print("Chi2 table = 5.991") 
print("")

print("p-value =", round(p,4))
print("")

print("Le Chi2 calculé est supérieur au Chi2 table, l'hypothèse H0 est donc rejetée, il y a donc une dépendance" +
      " entre les variables sexe et catégorie statistiquement significative (puisque p-value-test) < α.")
```

#### Le test V de Cramer

```python
# Il permet de comparer l'intensité du lien entre deux variables étudiées. 

# La formule de son calcul est la suivante : 
```

\begin{equation*}
V_{gof} = \sqrt{\frac{\chi^2}{n \times dof}}
\end{equation*}

```python
crosstab, test_results, expected = rp.crosstab(joinWO["sex"], joinWO["categ"],
                                               test= "chi-square",
                                               expected_freqs=True,
                                               prop= "cell")
test_results
```

#### Conclusion

```python
CramerV = test_results["results"][2]

if CramerV < .10:
    qual = 'négligeable'
elif CramerV < .20:
    qual = 'faible'
elif CramerV < .40:
    qual = 'modérée'
elif CramerV < .60:
    qual = 'relativement forte'
elif CramerV < .80:
    qual = 'forte'
else:
    qual = 'très forte' 
    
print("L'intensité de la corrélation entre les variables sexe et catégorie est donc :", qual, ".")
```

```python
# Dans une optique pragmatique, l’une des conceptions associées à la signification statistique est que l’on confond
# signification et importance. Ainsi, un résultat statistiquement significatif ne se traduit pas nécessairement 
# par une importance au niveau pratique. Il faut ici considérer qu’un échantillon suffisamment vaste mènera 
# généralement au rejet de l’hypothèse nulle ou, en d’autres mots, à un résultat statistiquement significatif et ce,
# peu importe la taille réelle de l’effet (pourvu qu’elle ne soit pas nulle).

# La taille de l’effet désigne à quel degré un phénomène donné est présent dans la population (Cohen 1988, p. 9, 
# traduction libre). Ainsi, une autre façon de concevoir l’hypothèse nulle est de dire que la taille de l’effet est
# nulle. Dans cette optique, la taille de l’effet décrit dans quelle mesure l’hypothèse nulle est fausse : plus la 
# taille de l’effet est grande, plus il est justifié de rejeter l’hypothèse nulle. Or, pour un grand échantillon, 
# un effet même minime suffira à la faire rejeter. Concrètement, ce problème peut être envisagé à travers l’exemple
# présenté plus haut : notre résultat significatif au test du khi-carré s’accompagne d’un effet (calculé à l’aide 
# du V de Cramer) considéré comme significatif, mais néanmoins négligeable, selon les critères de Cohen (V = 0,02 ;
# p = 0,00). La portée pratique d’une telle conclusion peut difficilement être vue comme importante. Néanmoins, 
# comme lors de l’interprétation de p, il faut éviter de porter un jugement mécanique sur la taille de l’effet 
# (Sawilowsky, 2003). Dans la mesure où les exigences du test d’hypothèses sont respectées, un résultat 
# statistiquement non significatif malgré une taille d’effet importante signifie que cet effet, aussi grand soit-il,
# pourrait être obtenu avec une probabilité supérieure à la valeur critère (Chow, 1996). Pour éviter une telle 
# situation, le chercheur doit veiller, notamment, à minimiser l’erreur de mesure. Cela dit, afin de fournir au 
# lecteur toute l’information requise pour juger des résultats obtenus, le chercheur devrait toujours fournir la 
# taille de l’effet au terme d’un test d’hypothèse (American Psychological Association, 2001 ; Wilkinson and the 
# Task Force on Statistical Inference, 1999). En fait, la taille de l’effet peut s’avérer plus utile à 
# l’interprétation des résultats que la valeur de p, surtout si l’ampleur de cet effet peut influencer les 
# décisions pratiques subséquentes (Kline, 2004).
```

#### Assumption Check

```python
# Pour pouvoir réaliser une ANOVA, il faut au préalable vérifier : 

#          - La variable quantitative suit une loi normale
#          - L'homogénéité des variances 

# Ici, on admet que ces conditions sont vérifiées pour pouvoir calculer η2. 
```

#### Hypothèse H0

$\left\{
\begin{array}{l}
 H_0\ : \ \bar{x}_1 = \bar{x}_2 = \bar{x}_k,\ les\ deux\ variables\ étudiées\ sont\ donc\ indépendantes \\
 H_1\ : \ \bar{x}_1 ≠ \bar{x}_2 ≠ \bar{x}_k,\ les\ deux\ variables\ sont\ donc\ dépendantes
\end{array}
\right.$

#### Test F et ANOVA

```python
corr = joinWO.groupby(['tranche_age', 'client_id'])['price'].sum().reset_index()

α = 0.05

aov = pg.anova(data=corr, dv='price', between='tranche_age', detailed=False)
aov
```

```python
if α > aov['p-unc'].item() : 
    
    print("La p-value étant inférieure au risque α, on rejette H0, les variables sont donc dépendantes.")
else :
    print("La p-value étant supérieur au risque α, H0 est vrai, les variables sont donc indépendantes. ")
```

#### Conclusion 

```python
# η2 permet de comparer l'intensité du lien entre deux variables étudiées.

# Interprétation η2 : 

# .01 ~ faible
# .06 ~ moyen
# >.14 ~ forte

# Par conséquent, la dépendance entre la variable tranche d'âge du client et le montant total des achats est 
# statistiquement significative, et la force de leur relation est plutôt moyenne. 
```
